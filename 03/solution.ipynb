```python
import numpy as np
import re

from operator import itemgetter
```

```python
def prepare_input(input):
    input.insert(0, '.' * len(input[0]))
    input.append('.' * len(input[0]))

    return [f'.{line}.' for line in input]
```

```python
def get_directions(length):
    directions = [
        (-1,1),
        (-1,0),
        (-1,-1)
    ]
    for i in range(length):
        directions.append((i,1))
        directions.append((i,-1))
    directions.append((length,1))
    directions.append((length,0))
    directions.append((length,-1))

    return directions

def get_field(length):
    directions = []
    for i in range(-length, length+1, 1):
        directions.append((i,-1))
    for i in range(-length, length+1, 1):
        directions.append((i,0))
    for i in range(-length, length+1, 1):
        directions.append((i,1))
    return directions
```

```python
def get_numbers(input, debug = False):
    input = prepare_input(input)
    nums = []

    for row, line in enumerate(input):
        numbers = re.findall(r'\d+', line)
        if len(numbers) == 0:
            continue
        start = 0
        if debug: print(f'...{row}...')
        for number in numbers:
            found = False
            start_search = line.find(number, start)
            for y,x in get_directions(len(number)):
                if found: 
                    continue
                if debug:
                    print(x, y, x+row, y+start, input[x+row][y+start_search])
                if re.match(r'[^\.\d]', input[x+row][y+start_search]) and not found:
                    found = True
                    if debug: print(number, input[x+row][y+start_search])
                    nums.append(int(number))
            start = start_search + len(number)

    return nums
```

```python
def get_valiate(input, debug = False):
    return sum(get_numbers(input, debug))
```

```python
def get_gears(input, debug=False):
    input = prepare_input(input)
    num_len = max([len(num) for num in re.findall(r'\d+', ''.join(input))])
    directions = get_field(num_len)
    sum = 0

    for row, line in enumerate(input):
        indexes =  [i for i, x in enumerate(line) if x == '*']
        for star in indexes:
            field = ''
            cur_x = directions[0][1]
            for y,x in directions:
                if x != cur_x: field += '\n'
                if 0 <= x+row < len(input) and 0 <= y+star < len(input[x+row]):
                    cur = input[x+row][y+star]
                    if re.match(r'\d', cur): field += cur
                    elif x == 0 and y == 0: field += '*'
                    else: field += '.'
                cur_x = x
            nums = get_numbers(field.strip().split('\n'))
            if debug: print(field, '\nNUMS:', nums, np.prod(nums))
            if len(nums) > 1:
                sum += np.prod(nums)
                if debug: print('\nSUMS:', sum)
    return sum
```
